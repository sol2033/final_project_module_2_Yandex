````python
# –ò–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –±–∏–±–ª–∏–æ—Ç–µ–∫ 
from sqlalchemy  import create_engine
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from phik import phik_matrix
from dotenv import load_dotenv 
import os 
````

__–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –æ—Ç —Ä–µ–≤—å—é–µ—Ä–∞ ‚Ññ1__

````python
# –ü–æ–¥–∫–ª—é—á–µ–Ω–∏–µ –∫ –±–∞–∑–µ –¥–∞–Ω–Ω—ã—Ö 
load_dotenv()
db_user = os.getenv('DB_USER')
db_password = os.getenv('DB_PWD')
db_host = os.getenv('DB_HOST')
db_port = os.getenv('DB_PORT')
db_name = os.getenv('DB_NAME')

db_config = {'user': db_user,
             'pwd': db_password, # –ø–∞—Ä–æ–ª—å
             'host': db_host,
             'port': db_port, # –ø–æ—Ä—Ç –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è
             'db': db_name # –Ω–∞–∑–≤–∞–Ω–∏–µ –±–∞–∑—ã –¥–∞–Ω–Ω—ã—Ö
             }

connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
)

engine = create_engine(connection_string)
````

````python
# –í—ã–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ –ë–î –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df_query = '''
WITH desktops_and_mobiles AS (
      SELECT *
      FROM afisha.purchases
      WHERE device_type_canonical IN ('mobile','desktop')
    )

SELECT t.user_id,
      t.device_type_canonical,
      t.order_id,
      t.created_dt_msk as order_dt,
      t.created_ts_msk as order_ts,
      t.currency_code,
      t.revenue,
      t.tickets_count,
      EXTRACT(DAY FROM (t.created_dt_msk - LAG(created_dt_msk) OVER (PARTITION BY user_id ORDER BY created_dt_msk))) AS days_since_prev,
       e.event_id,
       e.event_name_code as event_name,
       e.event_type_main,
       t.service_name,
       r.region_name,
       c.city_name
FROM afisha.purchases AS t
JOIN afisha.events AS e USING (event_id) 
JOIN afisha.city AS c ON c.city_id = e.city_id
JOIN afisha.regions AS r ON c.region_id = r.region_id
WHERE e.event_type_main != '—Ñ–∏–ª—å–º'
      AND device_type_canonical IN ('mobile','desktop')
ORDER BY t.user_id
'''
df = pd.read_sql_query(df_query, con = engine)
display(df.head(5))
````

````python
# –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –ø—Ä–æ—Ñ–∏–ª—è –∏ —á–∏—Å–ª–æ–º –∑–∞–∫–∞–∑–æ–º
corr_matrix = user_profile.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_matrix = corr_matrix[['total_orders']]
corr_matrix
````

````python
# –¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –¥–ª—è –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã
plt.figure(figsize = (10,6))
sns.heatmap(data = corr_matrix,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ total_orders —Å –¥—Ä—É–≥–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –ø—Ä–æ—Ñ–∏–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è')
plt.show()
````

````python
# –°–µ–≥–º–µ–Ω—Ç—ã –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏ –∏ 5+, –∏—Ö —Ç–µ–ø–ª–æ–≤—ã–µ –∫–∞—Ä—Ç—ã
users_2_4 = user_profile[(user_profile['is_two'] == 1) & (user_profile['is_five'] == 0)]
users_5_plus = user_profile[user_profile['is_five'] == 1]

corr_users_2_4 = users_2_4.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_users_2_4 = corr_users_2_4[['total_orders']]

corr_users_5_plus = users_5_plus.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_users_5_plus = corr_users_5_plus[['total_orders']]

print('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏')
display(corr_users_2_4)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 5+ –∑–∞–∫–∞–∑–∞–º–∏')
display(corr_users_5_plus)

plt.figure(figsize = (10,6))
sns.heatmap(data = corr_users_2_4,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ total_orders –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 2-4 –∑–∞–∫–∞–∑–∞–º–∏')
plt.show()

plt.figure(figsize = (10,6))
sns.heatmap(data = corr_users_5_plus,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ total_orders –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π —Å 5+ –∑–∞–∫–∞–∑–∞–º–∏')
plt.show()
````
