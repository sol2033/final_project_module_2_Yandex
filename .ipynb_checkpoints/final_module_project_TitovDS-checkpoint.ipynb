```python
# Подключение к базе данных 
db_config = {'user': 'praktikum_student', # имя пользователя
             'pwd': 'Sdf4$2;d-d30pp', # пароль
             'host': 'rc1b-wcoijxj3yxfsf3fs.mdb.yandexcloud.net',
             'port': 6432, # порт подключения
             'db': 'data-analyst-afisha' # название базы данных
             }

connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
)

engine = create_engine(connection_string)
```

```python
# Выгрузка данных из БД в датафрейм
df_query = '''
WITH desktops_and_mobiles AS (
      SELECT *
      FROM afisha.purchases
      WHERE device_type_canonical IN ('mobile','desktop')
    )

SELECT t.user_id,
      t.device_type_canonical,
      t.order_id,
      t.created_dt_msk as order_dt,
      t.created_ts_msk as order_ts,
      t.currency_code,
      t.revenue,
      t.tickets_count,
      EXTRACT(DAY FROM (t.created_dt_msk - LAG(created_dt_msk) OVER (PARTITION BY user_id ORDER BY created_dt_msk))) AS days_since_prev,
       e.event_id,
       e.event_name_code as event_name,
       e.event_type_main,
       t.service_name,
       r.region_name,
       c.city_name
FROM afisha.purchases AS t
JOIN afisha.events AS e USING (event_id) 
JOIN afisha.city AS c ON c.city_id = e.city_id
JOIN afisha.regions AS r ON c.region_id = r.region_id
WHERE e.event_type_main != 'фильм'
      AND device_type_canonical IN ('mobile','desktop')
ORDER BY t.user_id
'''
df = pd.read_sql_query(df_query, con = engine)
display(df.head(5))
```

Для оптимизации таблицы после перевода тенге в рубли были удалены также столбцы currency_code, revenue в виду неактуальности (после перевода все суммы в рублях и нет смысла хранить старый столбец суммы в тенге/рублях и исходную валюту, так как в дальнейшем анализе мы используем только рубли)

### 3. Создание профиля пользователя

В будущем отдел маркетинга планирует создать модель для прогнозирования возврата пользователей. Поэтому сейчас они просят вас построить агрегированные признаки, описывающие поведение и профиль каждого пользователя.

**Задача 3.1.** Постройте профиль пользователя — для каждого пользователя найдите:

- дату первого и последнего заказа;
- устройство, с которого был сделан первый заказ;
- регион, в котором был сделан первый заказ;
- билетного партнёра, к которому обращались при первом заказе;

- общее количество заказов;
- средняя выручка с одного заказа в рублях;
- среднее количество билетов в заказе;
- среднее время между заказами.

После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


```python
# Создание датафрейма профиля пользователя
user_profile = pd.DataFrame()
user_profile = df.sort_values(['user_id','order_dt']).groupby('user_id').agg(
    first_order_dt=('order_dt', 'min'),
    last_order_dt=('order_dt', 'max'),
    first_device=('device_type_canonical', 'first'),
    first_region=('region_name', 'first'),
    first_service=('service_name', 'first'),
    first_event_type=('event_type_main', 'first'),
    total_orders=('order_id', 'count'),
    avg_revenue_rub=('revenue_rub', 'mean'),
    avg_tickets_count=('tickets_count', 'mean'),
    avg_days_between=('days_since_prev', 'mean')
).reset_index()

user_profile['avg_revenue_rub'] = user_profile['avg_revenue_rub'].round(2)
user_profile['avg_tickets_count'] = user_profile['avg_tickets_count'].round(2)
user_profile['avg_days_between'] = user_profile['avg_days_between'].round(2)

user_profile['is_two'] = (user_profile['total_orders'] >= 2).astype('int')
user_profile['is_five'] = (user_profile['total_orders'] >= 5).astype('int')

display(user_profile)
```

При группировке по first_event_type пользователи распределены неравномерно. Преобладают концерты (43%), другое (24%) и театры (21%), в остальных пользователей 5% от общего числа и менее

```python
# Группировка пользователей по сегментам
def segments_info(row):
        print(f'Группировка по {row}')
        grouped_df = user_profile.groupby(row)["user_id"].agg("count").reset_index().sort_values(by = 'user_id', ascending = False)
        grouped_df.columns = ['row','user_counts']
        grouped_df['share'] = round(grouped_df['user_counts'] / grouped_df['user_counts'].sum(),2)
        display(grouped_df)
        return grouped_df
        
grouped_event_df = segments_info('first_event_type')
grouped_device_df = segments_info('first_device')
grouped_region_df = segments_info('first_region')
grouped_service_df = segments_info('first_service')
```

```python
#Датафрейм с данными о кол-ве пользователей и доли с повторными заказами для каждого дня недели
day_of_week_df = user_profile.groupby(user_profile['first_order_dt'].dt.weekday+1).agg(
                                                                                        {
                                                                                            'user_id':'count',
                                                                                            'is_two':'sum'
                                                                                        }
)
day_of_week_df = day_of_week_df.reset_index()
day_of_week_df['day_week'] = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday']
day_of_week_df.columns = ['week_num','user_counts','with_repeat_order_share','day_week']
day_of_week_df = day_of_week_df.drop(columns = 'week_num')
day_of_week_df = day_of_week_df.set_index('day_week')
day_of_week_df['with_repeat_order_share'] = round(day_of_week_df['with_repeat_order_share'] / day_of_week_df['user_counts'],2)
day_of_week_df
```

```python
# Визуализация датафрейма day_of_week_df
day_of_week_df['user_counts'].sort_values(ascending = False).plot(
                                    kind = 'bar',
                                    xlabel = 'Дни недели',
                                    ylabel = 'Кол-во пользователей',
                                    title = 'Распределение пользователей по дням недели их первого заказа',
                                    figsize = (10,6)
                                    )
plt.show()

day_of_week_df['with_repeat_order_share'].sort_values(ascending = False).plot(
                                    kind = 'bar',
                                    xlabel = 'Дни недели',
                                    ylabel = 'Доля пользователей',
                                    title = 'Доля пользователей с повторными заказами по дню недели их первого заказа',
                                    figsize = (10,6)
                                    )
plt.show()
```

#### 4.4. Корреляционный анализ количества покупок и признаков пользователя

Изучите, какие характеристики первого заказа и профиля пользователя могут быть связаны с числом покупок. Для этого используйте универсальный коэффициент корреляции `phi_k`, который позволяет анализировать как числовые, так и категориальные признаки.

---

**Задача 4.4.1:** Проведите корреляционный анализ:
- Рассчитайте коэффициент корреляции `phi_k` между признаками профиля пользователя и числом заказов (`total_orders`). При необходимости используйте параметр `interval_cols` для определения интервальных данных.
- Проанализируйте полученные результаты. Если полученные значения будут близки к нулю, проверьте разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.
- Ответьте на вопрос: какие признаки наиболее связаны с количеством заказов?

---

Ранее при предобработке данных в total_orders были исключены выбросы. В матрице корреляции не все признаки стремятся к нулю, поэтому необходимости выделять отдельные сегменты пользователей нет. На основе матрицы и тепловой карты можно отметить, что с количеством заказов наиболее связаны даты первого и последнего заказа (по 0,42), средний промежуток между заказами(0,3) и меньшую корреляцию имеют среднее количество билетов и средняя сумма чеков (по 0,23), первый регион (0,12). Корреляция практически отсутствует по таким признакам, как устройства, типы мероприятий, сервис по продаже билетов, которые были выбраны в первых заказах, что выглядит логичным 

При рассмотрении отдельных сегментов пользователей (2-4 заказов и 5+) можно отметить, что корреляция с first_event_type, first_service, first_device еще меньше, чем при обработке всех сегментов сразу. У пользователей с 5+ заказами повышенная корреляция avg_days_between с total_orders (0.70) по сравнению с общей матрицей корреляции. У сегмента пользователей с 2-4 заказами наблюдается повышенная относительно общей матрицы корреляция с total_orders столбцом avg_tickets_count (0,37). У обоих сегментов корреляция first_order_dt и last_order_dt меньше, чем в общей матрице. 

Следовательно, при рассмотрении всех сегментов сразу наблюдается связь с датами первого и последнего заказов, средним промежутком, средним кол-вом билетом (малая корреляция), средней суммой чеков (аналогично малая). В сегменте с 2-4 заказами с total_orders имеют связь среднее кол-во билетов, средним кол-вом дней, датами первого и последнего заказов. В сегменте 5+ заказов с total_orders имеют связь среднее кол-во дней между заказами, первым регионом, средним кол-вом билетов, средней стоимостью билетов, датами первого и последнего заказов

```python
# Матрица корреляции между признаками профиля и числом заказом
corr_matrix = user_profile.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_matrix = corr_matrix[['total_orders']]
corr_matrix
```

```python
# Тепловая карта для корреляционной матрицы
plt.figure(figsize = (10,6))
sns.heatmap(data = corr_matrix,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('Тепловая карта корреляции total_orders с другими признаками профиля пользователя')
plt.show()
```

```python
# Сегменты пользователей с 2-4 заказами и 5+, их тепловые карты
users_2_4 = user_profile[(user_profile['is_two'] == 1) & (user_profile['is_five'] == 0)]
users_5_plus = user_profile[user_profile['is_five'] == 1]

corr_users_2_4 = users_2_4.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_users_2_4 = corr_users_2_4[['total_orders']]

corr_users_5_plus = users_5_plus.drop(columns = ['user_id','is_two','is_five']).phik_matrix(interval_cols = ['total_orders','avg_revenue_rub','avg_tickets_count','avg_days_between'])
corr_users_5_plus = corr_users_5_plus[['total_orders']]

print('Матрица корреляции для пользователей с 2-4 заказами')
display(corr_users_2_4)
print('Матрица корреляции для пользователей с 5+ заказами')
display(corr_users_5_plus)

plt.figure(figsize = (10,6))
sns.heatmap(data = corr_users_2_4,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('Тепловая карта корреляции total_orders для пользователей с 2-4 заказами')
plt.show()

plt.figure(figsize = (10,6))
sns.heatmap(data = corr_users_5_plus,
            annot = True,
            fmt = '.2f',
            cmap = 'coolwarm')
plt.title('Тепловая карта корреляции total_orders для пользователей с 5+ заказами')
plt.show()
```
